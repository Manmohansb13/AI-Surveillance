# Restricted Zone

```python
from ultralytics import YOLO
from deep_sort_realtime.deepsort_tracker import DeepSort
import cv2
import torch

# ------------------ 1. Device ------------------
device = "mps" if torch.backends.mps.is_available() else "cpu"
print("Using device:", device)

# ------------------ 2. Load YOLOv8n ------------------
model = YOLO("yolov8n.pt")  # Nano model for speed

# ------------------ 3. Initialize DeepSORT ------------------
tracker = DeepSort(max_age=30)

# ------------------ 4. Input and output ------------------
video_path = "crowd.mp4"
output_path = "working_Crowd_restricted1.mp4"

cap = cv2.VideoCapture(video_path)
width  = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps    = int(cap.get(cv2.CAP_PROP_FPS))

# VideoWriter to save output
out = cv2.VideoWriter(output_path,
                      cv2.VideoWriter_fourcc(*'mp4v'),
                      fps, (width, height))

# ------------------ 5. Define restricted zone ------------------
zone_w, zone_h = width // 3, height // 3  # size of the zone
zone_x1 = width // 2 - zone_w // 2
zone_y1 = height // 2 - zone_h // 2
zone_x2 = zone_x1 + zone_w
zone_y2 = zone_y1 + zone_h

# Dictionary to track time each person spends inside the restricted zone
person_time_in_zone = {}

# ------------------ 6. Process only first 50 frames ------------------
frame_count = 0
for result in model(video_path, device=device, stream=True):
    frame_count += 1
    if frame_count > 150:
        break

    frame = result.orig_img  # BGR frame

    # Draw restricted zone
    cv2.rectangle(frame, (zone_x1, zone_y1), (zone_x2, zone_y2), (0, 0, 255), 2)
    cv2.putText(frame, "RESTRICTED", (zone_x1 + 5, zone_y1 - 10),
                cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 0, 255), 2)

    # ------------------ 7. Extract person detections ------------------
    detections = []
    for box in result.boxes:
        cls = int(box.cls[0].item())
        conf = float(box.conf[0].item())
        x1, y1, x2, y2 = map(int, box.xyxy[0].tolist())
        w, h = x2 - x1, y2 - y1
        if cls == 0 and w > 10 and h > 10:  # person class
            detections.append(([x1, y1, w, h], conf, cls))

    # ------------------ 8. Update DeepSORT tracker ------------------
    if detections:
        tracks = tracker.update_tracks(detections, frame=frame)

        for track in tracks:
            if not track.is_confirmed():
                continue

            x1, y1, x2, y2 = map(int, track.to_ltrb())
            cx = (x1 + x2) // 2  # center x of the person
            cy = (y1 + y2) // 2  # center y

            # Draw tracking box
            cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)

            # Check if person is inside restricted zone
            if zone_x1 <= cx <= zone_x2 and zone_y1 <= cy <= zone_y2:
                person_time_in_zone[track.track_id] = person_time_in_zone.get(track.track_id, 0) + 1

            # Display ID and time in zone (seconds)
            time_in_sec = person_time_in_zone.get(track.track_id, 0) / fps
            cv2.putText(frame, f"ID {track.track_id} | {time_in_sec:.1f}s", 
                        (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 0), 2)

    # Write frame to output video
    out.write(frame)

cap.release()
out.release()
print(f"Processing complete! Saved tracked video as {output_path}")
```
